```python
mode_value = df['Mental_Health_Condition'].mode()
if not mode_value.empty:
    mode_value = mode_value[0] 
else:
    mode_value = "Unknown"  
```

```python
# Replace missing values with the mode
mode_value = df['Mental_Health_Condition'].mode()[0]  
df['Mental_Health_Condition'] = df['Mental_Health_Condition'].fillna(mode_value)

print(df['Mental_Health_Condition'].head())
```

```python
# Reset the index to avoid slicing issues
df.reset_index(drop=True, inplace=True)
print(df.head())
```

```python
# Safely calculate and fill missing values in a column
if 'Mental_Health_Condition' in df.columns:
    mode_value = df['Mental_Health_Condition'].mode()[0] 
    df['Mental_Health_Condition'] = df['Mental_Health_Condition'].fillna(mode_value)
else:
    print("'Mental_Health_Condition' column does not exist in the DataFrame.")
```

```python

# Save the DataFrame to a CSV file
df.to_csv("updated_dataset.csv", index=False)

# Inform the user where the file is saved
print("The updated dataset has been saved as 'updated_dataset.csv' in the current directory.")
```

```python
sleep_mapping = {"Good": 1, "Average": 2, "poor": 3}
df["Sleep_Quality"] = df["Sleep_Quality"].map(sleep_mapping)
```

```python
sleep_mapping  = df.groupby("Access_to_Mental_Health_Resources")["Sleep_Quality"].mean()
sleep_mapping 
```

```python
avg_years_worked = df.groupby("Years_of_Experience")["Sleep_Quality"].mean()
avg_years_worked.head() 
```

```python
stress_mapping = {"Low": 1, "Medium": 2, "High": 3}
df["Stress_Level"] = df["Stress_Level"].map(stress_mapping)
avg_stress = df.groupby("Work_Location")["Stress_Level"].mean()
avg_stress
```

```python
#aggregate data:calculate average and total work done by work hours and job role
agg_Work_Location = df.groupby("Work_Location").agg(Avg_Stress=("Stress_Level","mean"),Total_Hours_Worked=("Hours_Worked_Per_Week","sum"))
agg_Work_Location                                                    
```

```python
sns.lineplot(x='Age', y='Years_of_Experience', data=df)
plt.title('Years Experince and Age')
plt.grid()
plt.show()
```

```python
gender_counts = df['Work_Location'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(gender_counts, labels=gender_counts.index,autopct='%1.1f%%', startangle=90, colors=['skyblue', 'red','green'])
plt.title('Work_location')
plt.show()
```

```python
# Job role distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, y='Job_Role', order=df['Job_Role'].value_counts().index,hue="Job_Role", palette='viridis')
plt.title('Job Role Distribution')
plt.show()
```

```python
sns.histplot(df['Hours_Worked_Per_Week'], bins=13, color='r')
plt.title('Distribution of Hours Worked Per Week')
plt.show()
```

```python
# Pie chart for gender distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(4, 6))
plt.pie(gender_counts, labels=gender_counts.index,autopct='%1.1f%%', startangle=90, colors=['skyblue', 'red'])
plt.title('Gender Distribution')
plt.show()
```

```python
sns.countplot(x=df['Industry'], hue='Gender', data=df)
plt.xticks(rotation=45)
plt.grid()
plt.title("Industry Specific Gender Coloring")
plt.show()
```

```python
# Create a pivot table for Number_of_Virtual_Meetings
meetings_pivot = df.pivot_table(index='Age', columns='Mental_Health_Condition', values='Number_of_Virtual_Meetings', aggfunc='mean')

# Heatmap for Number_of_Virtual_Meetings
plt.figure(figsize=(8, 6))
sns.heatmap(meetings_pivot, annot=True, cmap='viridis', fmt=".1f", cbar_kws={'label': 'Number of Virtual Meetings'})
plt.show()
```

```python
# Boxplot for Years of Experience vs. Satisfaction with Remote Work
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Years_of_Experience', y='Satisfaction_with_Remote_Work',hue='Satisfaction_with_Remote_Work', palette='pastel')
plt.title('Satisfaction with Remote Work by Years of Experience')
plt.show()
```

```python
# Bar plot for Work Location vs. Productivity Change
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Work_Location', hue='Productivity_Change', palette='Set2')
plt.title('Work Location vs. Productivity Change')
plt.legend(title='Productivity Change')
plt.show()
```

```python
# Pie chart for Access to Mental Health Resources
plt.figure(figsize=(4, 4))
df['Access_to_Mental_Health_Resources'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'pink'])
plt.title('Access to Mental Health Resources')
plt.show()
```

```python
# Scatter plot for Number of Virtual Meetings vs. Satisfaction with Remote Work
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Number_of_Virtual_Meetings', y='Satisfaction_with_Remote_Work', hue='Gender', palette='deep')
plt.title('Number of Virtual Meetings vs. Satisfaction with Remote Work')
plt.show()
```
